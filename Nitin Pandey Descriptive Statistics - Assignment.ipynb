```python
sns.boxplot(y = df["total_bill"])
#Total bills value above 40 are outliers
#Rightly skewed as whisker stretches on right
#25 percentile - 13
#50 percentile - 18
#75 percentile - 24
#IQR = 11, 6 outliers
```

```python
df.total_bill.describe()
#Validates the above observation
```

```python
l1 = list(df["total_bill"])
l1
```

```python
l2 = []
l3 = []
for i in l1:
    z = (i - 19.785943)/8.902412
    l2.append(z)
print("z score of total bill is ", l2)
for j in l2:
    if j < -3 or j > 3:
        l3.append(j)
print("Outliers are ", l3)  
#Code is showing 4 outliers, however IQR is showing 6 outliers, IQR is correctly showing outliers in this case as it is skewed
```

```python
sns.boxplot(y = df["tip"])
#Tip value above 6 are outliers
#Rightly skewed as whisker stretches on right
#25 percentile - 2
#50 percentile - 3
#75 percentile - 3.5
#IQR = 1.5, 6 outliers
```
